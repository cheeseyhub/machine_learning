```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split

data_file = pd.read_csv("./car_data.csv");
data_file = pd.get_dummies(data_file,columns=['Brand']);


Mileage_column_mean = data_file['Mileage'].mean();
Price_column_mean = data_file['Price'].mean();

data_file['Mileage'] =data_file['Mileage'].fillna(Mileage_column_mean);
data_file['Price'] = data_file['Price'].fillna(Price_column_mean);

data_file['Buy'] = data_file['Buy'].replace({'Yes':1,'No':0});


X = data_file.drop('Buy',axis=1);
Y = data_file['Buy'];

X_train,X_test,y_train,y_test = train_test_split(
    X,Y,test_size=0.2,random_state=42
)

print(X_train.shape)
print(X_test.shape)
print(X_train.head())

```
